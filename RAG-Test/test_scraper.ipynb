```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

url = "https://kohplanner.com/food/10-classic-thai-dishes-recipes/"

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î Strainer: ‡πÄ‡∏≠‡∏≤‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤‡∏ö‡∏ó‡∏Ñ‡∏ß‡∏≤‡∏° (‡∏ï‡∏±‡∏î Header/Footer ‡∏ó‡∏¥‡πâ‡∏á)
# ‡∏à‡∏≤‡∏Å HTML ‡∏ó‡∏µ‡πà‡∏Ñ‡∏∏‡∏ì‡∏™‡πà‡∏á‡∏°‡∏≤ ‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô class="entry-content"
bs4_strainer = bs4.SoupStrainer(class_="entry-content")

loader = WebBaseLoader(
    web_paths=(url,), 
    bs_kwargs={"parse_only": bs4_strainer}
)

print(f"üîÑ Loading from: {url}...")
try:
    docs = loader.load()
    print(f"‚úÖ Loaded Successfully! ({len(docs)} document)")
    
    # ‡πÇ‡∏ä‡∏ß‡πå‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤ 1000 ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡πÅ‡∏£‡∏Å
    content = docs[0].page_content
    print("\n--- Sample Content ---")
    print(content[:1000]) 
    print("\n(‡∏¢‡∏±‡∏á‡∏°‡∏µ‡∏ï‡πà‡∏≠...)")
except Exception as e:
    print(f"‚ùå Error: {e}")
```
