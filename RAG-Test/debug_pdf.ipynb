```python
import os
from langchain_community.document_loaders import PyPDFLoader, PyPDFDirectoryLoader

# 1. ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡πÑ‡∏ü‡∏•‡πå‡∏°‡∏µ‡∏≠‡∏¢‡∏π‡πà‡∏à‡∏£‡∏¥‡∏á‡πÑ‡∏´‡∏° (List Files)
target_dir = os.path.abspath("data/recipes")
print(f"üìÇ Target Directory: {target_dir}")

if os.path.exists(target_dir):
    files = os.listdir(target_dir)
    pdf_files = [f for f in files if f.lower().endswith('.pdf')]
    print(f"üìÑ PDF Files Found ({len(pdf_files)}): {pdf_files}")
else:
    print("‚ùå Directory does not exist!")

# 2. ‡∏•‡∏≠‡∏á‡∏≠‡πà‡∏≤‡∏ô‡πÅ‡∏ö‡∏ö‡∏£‡∏∞‡∏ö‡∏∏‡πÑ‡∏ü‡∏•‡πå‡∏ï‡∏£‡∏á‡πÜ (Direct Load)
if pdf_files:
    first_pdf_path = os.path.join(target_dir, pdf_files[0])
    print(f"\nüîÑ Attempting to load: {first_pdf_path}")
    try:
        loader = PyPDFLoader(first_pdf_path)
        docs = loader.load()
        print(f"‚úÖ Success! Loaded {len(docs)} pages from single file.")
    except Exception as e:
        print(f"‚ùå Failed to load single file: {e}")

# 3. ‡∏•‡∏≠‡∏á‡∏≠‡πà‡∏≤‡∏ô‡∏ó‡∏±‡πâ‡∏á‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå (Directory Load)
print("\nüîÑ Attempting directory load...")
try:
    dir_loader = PyPDFDirectoryLoader(target_dir)
    dir_docs = dir_loader.load()
    print(f"‚úÖ Directory Loader Result: {len(dir_docs)} pages")
except Exception as e:
    print(f"‚ùå Directory Loader Failed: {e}")
```
